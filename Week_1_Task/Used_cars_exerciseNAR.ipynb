## The dataset contains information about Used Cars Price with the following columns:

`index`: Index

`Name`: The brand and model of the car.

`Location`: The location in which the car is being sold or is available for purchase.

`Year`: The year or edition of the model.

`Kilometers_Driven`: The total kilometres driven in the car by the previous owner(s) in KM.

`Fuel_Type`: The type of fuel used by the car. (Petrol / Diesel / Electric / CNG / LPG)

`Transmission`: The type of transmission used by the car. (Automatic / Manual)

`Owner_Type`: Whether the ownership is Firsthand, Second hand or other.

`Mileage`: The standard mileage offered by the car company in kmpl or km/kg

`Engine`: The displacement volume of the engine in cc.

### Read the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('Used_Cars.csv')
```

```python
df = df.drop(columns=['Unnamed: 0'], axis = 1)
df.info()
```

### Ensure data types are appropriate for each column.

---

```python
def extract_numeric(value):
    if isinstance(value, str):
        numeric_str = ''.join(c for c in value if c.isdigit() or c == '.')
        try:
            return float(numeric_str)
        except ValueError:
            return None
    return value

df['Mileage'] = df['Mileage'].apply(extract_numeric)
df['Power'] = df['Power'].apply(extract_numeric)
df['New_Price'] = df['New_Price'].apply(extract_numeric)
df['Engine'] = df['Engine'].apply(extract_numeric)
df.info()
df.tail()
```

### Handling Missing Values

```python
df['Engine'].fillna(df['Engine'].mode()[0], inplace=True)
df.info()
```

```python
df = df.drop(columns=['New_Price'], axis = 1)
df.info()
```

```python
df['Seats'].fillna(df['Seats'].mode()[0], inplace=True)
df.info()
```

```python
mean_Seats = df['Power'].mean()
df['Power'].fillna(mean_Seats, inplace=True)
df.info()
```

```python
df['Mileage'].fillna(df['Mileage'].mode()[0], inplace=True)
df.info()
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df = remove_outliers(df, 'Kilometers_Driven')
df = remove_outliers(df, 'Mileage')
df = remove_outliers(df, 'Engine')
df = remove_outliers(df, 'Power')
df = remove_outliers(df, 'Seats')
df = remove_outliers(df, 'Price')
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = StandardScaler()
df['Kilometers_Driven'] = scaler.fit_transform(df[['Kilometers_Driven']])
df['Engine'] = scaler.fit_transform(df[['Engine']])
df['Power'] = scaler.fit_transform(df[['Power']])
```

```python
from sklearn.preprocessing import LabelEncoder

label_E = LabelEncoder()
df['Year'] = label_E.fit_transform(df['Year'])
df['Seats'] = label_E.fit_transform(df['Seats'])
```
