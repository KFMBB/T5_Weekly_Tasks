## The dataset contains information about Used Cars Price with the following columns:

`index`: Index

`Name`: The brand and model of the car.

`Year`: The year or edition of the model.

`Owner_Type`: Whether the ownership is Firsthand, Second hand or other.

`Mileage`: The standard mileage offered by the car company in kmpl or km/kg

`Engine`: The displacement volume of the engine in cc.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('Used_Cars.csv')
```

```python
# Convert Seats column to Integer
df['Seats'] = df['Seats'].astype(str)

# Convert Year column to datetime
df['Year'] = df['Year'].astype(str)
```

```python
# Function to clean the mileage column
def clean_data(column):
        return float(column.split()[0])
    

# Apply the function to the Mileage column
df['Mileage'] = df['Mileage'].apply(clean_data)



# Apply the function to the Engine column
df['Engine'] = df['Engine'].apply(clean_data)


# Apply the function to the Power column
# To DO: error could not convert string to float: Null
df['Power'] = df['Power'].str.replace('null' , '0')
df['Power'] = df['Power'].apply(clean_data)


df.head()
```

```python
# Function to remove outliers using the IQR method
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers from 'Fare' using the remove_outilers function (1 line)
df = remove_outliers(df, "Power")
df = remove_outliers(df, "Engine")
df = remove_outliers(df, "Price")
df = remove_outliers(df, "Mileage")
```
